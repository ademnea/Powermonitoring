```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def process_df(filepath):
    df = pd.read_csv(filepath, skiprows=2)
    
    df.columns = ['Voltage', 'Current']
    # drop negative values of current
    df = df[df['Current'] > 0]
    df['Power'] = df['Voltage'] * df['Current']
    
    return df.dropna()

def plot_variables(df, var2):
    
    mean_var2 = df[var2].mean()
    
    fig, ax = plt.subplots(figsize=(14, 6))
    ax.plot(df.index, df[var2], label=var2, color='blue', linestyle='-')
    
    ax.set_xlabel('Time')
    ax.set_ylabel(var2)
    ax.set_title(f'{var2} Over Time')
    
    ax.axhline(mean_var2, color='red', linestyle='--', label=f'Mean {var2}')
    ax.legend()
    
    plt.show()
```

```python
df = process_df('C:/Users/mukiibirogers/OneDrive/Documents/Powermonitoring/arduino_data/data01.csv')
df2 = process_df('C:/Users/mukiibirogers/OneDrive/Documents/Powermonitoring/arduino_data/data02.csv')
df.head()
```

```python
plot_variables(df, 'Power')
```
