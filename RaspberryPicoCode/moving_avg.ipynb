```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def process_df(filepath):
    df = pd.read_csv(filepath)
    
    df['current'] = (df['average_voltage'] - 2.3) / 0.185
    df['power'] = df['current'] * df['average_voltage']
    
    return df

def plot_variables(df, time_col, var2):

    mean_var2 = df[var2].mean()
    
    fig, ax = plt.subplots(figsize=(14, 6))
    ax.plot(df[time_col], df[var2], label=var2, color='blue')
    ax.axhline(mean_var2, color='red', linestyle='--', label=f"Mean {var2}: {mean_var2:.2f}")
    ax.set_xlabel(time_col)
    ax.set_ylabel(var2)
    ax.legend()
    plt.show()
    
```

```python
df8 = process_df("C:/Users/mukiibirogers/OneDrive/Documents/Powermonitoring/data/data08.csv")
df9 = process_df("C:/Users/mukiibirogers/OneDrive/Documents/Powermonitoring/data/data09.csv") # test1 of 0 current
df10 = process_df("C:/Users/mukiibirogers/OneDrive/Documents/Powermonitoring/data/data10.csv") # test2 of 0 current
df11 = process_df("C:/Users/mukiibirogers/OneDrive/Documents/Powermonitoring/data/data11.csv") # increasing context window from 10 to 100
df12 = process_df("C:/Users/mukiibirogers/OneDrive/Documents/Powermonitoring/data/data12.csv") # increasing context window from 100 to 150
df13 = process_df("C:/Users/mukiibirogers/OneDrive/Documents/Powermonitoring/data/data13.csv") # running for 10 seconds
df14 = process_df("C:/Users/mukiibirogers/OneDrive/Documents/Powermonitoring/data/data14.csv") # running for 2 minutes

print(f"df8: {df8.shape}, df9: {df9.shape}, df10: {df10.shape}, df11: {df11.shape}, df12: {df12.shape}, df13: {df13.shape}, df14: {df14.shape}")
```
