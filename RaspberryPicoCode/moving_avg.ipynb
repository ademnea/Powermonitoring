```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def process_df(filepath):
    df = pd.read_csv(filepath)
    
    df['current'] = (df['average_voltage'] - 2.455) / 0.185
    df['power'] = df['current'] * df['average_voltage']
    
    return df

def plot_variables(df, time_col, var2):

    mean_var2 = df[var2].mean()
    
    fig, ax = plt.subplots(figsize=(14, 6))
    ax.plot(df[time_col], df[var2], label=var2, color='blue')
    ax.axhline(mean_var2, color='red', linestyle='--', label=f"Mean {var2}")
    ax.set_xlabel(time_col)
    ax.set_ylabel(var2)
    ax.legend()
    plt.show()
    
```

```python
df8 = process_df("C:/Users/mukiibirogers/OneDrive/Documents/Powermonitoring/data/data08.csv")

print(f"df8: {df8.shape}")
```

```python
plot_variables(df8, 'time', 'average_voltage')
```

```python
plot_variables(df8, 'time', 'power')
```
