```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def process_df(filepath):
    df = pd.read_csv(filepath)
    
    df['current'] = (df['average_voltage'] - 2.4075) / 0.185
    df['power'] = df['current'] * df['average_voltage']
    
    return df

def plot_variables(df, time_col, var2):

    mean_var2 = df[var2].mean()
    
    fig, ax = plt.subplots(figsize=(14, 6))
    ax.plot(df[time_col], df[var2], label=var2, color='blue')
    ax.axhline(mean_var2, color='red', linestyle='--', label=f"Mean {var2}: {mean_var2:.2f}")
    ax.set_xlabel(time_col)
    ax.set_ylabel(var2)
    ax.legend()
    ax.grid()
    # plt.xticks(np.arange(0, 10))
    plt.show()
    
```
